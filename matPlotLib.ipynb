## Subplots

* Subplot(row, column, index)

```python
import matplotlib.pyplot as plt
import numpy as np

# 1 Sin(x)
plt.subplot(2, 2, 1)
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y, marker = 'o', mec = 'r', mfc = 'b', ms = 10, ls = 'dotted', lw = 2, c = 'k')
f1 = {"family": "serif", "size": 7, "color": 'r'}
f2 = {"family": "fantasy", "size": 7, "color": 'r'}
plt.xlabel("x", fontdict = f1)
plt.ylabel("y", fontdict = f2)
plt.title("Sin(x)", fontdict = f2)


# 2 Cos(x)
plt.subplot(2, 2, 2)
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.cos(x)
plt.plot(x, y, marker = 'p', mec = 'g', mfc = 'k', ms = 10, ls = "dashdot", lw = 3, c = 'r')
f1 = {"family": "Arial", "size": 8, "color": "y"}
plt.xlabel("x", fontdict = f1)
plt.ylabel("y", fontdict = f1)
plt.title("Cos(x)", fontdict = f2)

# 3 tan(x)
plt.subplot(2, 2, 3)
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.tan(x)
plt.plot(x, y, marker = 'h', mec = 'b', mfc = 'y', ms = 1, ls = 'dashed', lw = 2, c = 'k')
plt.xlabel("x", fontdict = f1)
plt.ylabel("y", fontdict = f1)
plt.title("Tan(x)", fontdict = f2)

# 4 sqrt(x)
plt.subplot(2, 2, 4)
x = np.linspace(0, 100, 1000)
y = np.sqrt(x)
plt.plot(x, y, marker = 'h', mec = 'y', mfc = 'r', ms = 10, ls = "solid", lw = 3, c = "b")
plt.xlabel("x", fontdict = f1)
plt.ylabel("y", fontdict = f1)
plt.title("Sqrt(x)", fontdict = f2)

plt.suptitle("All trig", fontdict = f2)
```

## Barplot

```python
dt = {"Apple": 40, "Banana": 25, "Mango": 80, "Grapes": 45, "Pineapple": 25}
categories = list(dt.keys())
values = list(dt.values())

plt.bar(categories, height = values)
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.title("Fruits prices")
plt.show()
```

```python
dt = {"Apple": 40, "Banana": 25, "Mango": 80, "Grapes": 45, "Pineapple": 25}
categories = list(dt.keys())
values = list(dt.values())

plt.barh(categories, width = values)
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.title("Fruits prices")
plt.show()
```

```python
fruits = ["Apple", "Banana", "Cherry", "Mango", "Pineapple"]
price_22 = [40, 25, 80, 45, 25]
price_21 = [35, 25, 70, 50, 20]

width = 0.5

x = np.linspace(0, 5, len(fruits))
plt.bar(x + width/2, price_22, width, color = "r", label = "2022")
plt.bar(x - width/2, price_21, width, color = "b", label = "2021")

plt.xticks(x, fruits)
plt.xlabel("Fruits")
plt.ylabel("Prices")
plt.title("Comparison of prices")
plt.legend()
plt.show()
```

## Pie chart

```python
covid_cases = {"AP": 1000, "TS": 1500, "UP": 3000, "MH": 5000}

labels = list(covid_cases.keys())
values = list(covid_cases.values())

plt.pie(values, labels = labels)
plt.show()
```

```python
covid_cases = {"AP": 1000, "TS": 1500, "UP": 3000, "MH": 5000}

labels = list(covid_cases.keys())
values = list(covid_cases.values())
exp = [0, 0.2, 0, 0]

plt.pie(values, labels = labels, explode = exp)
plt.legend()
plt.show()
```
