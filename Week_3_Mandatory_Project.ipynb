```python
#5.Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
 
def calculate_median(data):
    data = np.sort(data)
    if len(data) % 2 == 0:
        # if the data set has an even length, return the average of the two middle values
        middle_index1 = len(data) // 2
        middle_index2 = middle_index1 - 1
        return (data[middle_index1] + data[middle_index2]) / 2
    else:
        # if the data set has an odd length, return the middle value
        middle_index = len(data) // 2
        return data[middle_index]




  

 
```

```python



# load the dataset
df = pd.read_excel('/content/Dataset.xlsx')
df

# create a new column 'total_bedroom_size' with default value 'unknown'
df['total_bedroom_size'] = 'unknown'

# assign values to 'total_bedroom_size' column based on conditions
df.loc[df['total_bedrooms'] <= 10, 'total_bedroom_size'] = 'small'
df.loc[(df['total_bedrooms'] > 10) & (df['total_bedrooms'] < 1000), 'total_bedroom_size'] = 'medium'
df.loc[df['total_bedrooms'] >= 1000, 'total_bedroom_size'] = 'large'

#save the final dataset
df.to_excel('/content/Dataset.xlsx', index=False)
df
```
